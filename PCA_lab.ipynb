## Linear Least Square Regression using Linear Algebra

### Example 1

#### Step 1: Dataset creation

```

import numpy as np
x = np.arange(-10,10,0.1)
x = x.reshape(len(x),1)
x = x+np.random.randn(x.shape[0],1)### Dataset -x values
m = 1
c = 2

y = m*x + c +   3* np.random.rand(x.shape[0],1)## Dataset - y values
```

#### Step 2: Plotting Dataset

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.rcParams['axes.facecolor'] = 'yellow'
plt.grid(c='grey')
plt.scatter(x, y)
plt.axis('equal')
# plt.xlabel('X')
# plt.ylabel('Y')
#plt.title('Positively Correlated')
plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)
plt.title('Data',fontsize=30)
#plt.grid()
#plt.legend(loc='upper right', fontsize=30)
plt.show()
```

#### Step 3: Concepts


\begin{equation}
\vec{y} = A\vec{x} + \vec{e}\\
A^{T}\vec{e} = \vec{0}\\
A^{T}(\vec{y} - A\vec{x}) = \vec{0}\\
\vec{x} = (A^{T}A)^{-1}A^{T}\vec{y}
\end{equation}



#### Step 4: Defining A matrix

```
A = np.append(x, np.ones((x.shape[0],1)),axis = 1)
```

#### Step 5: Implementation of Linear Least Square Method 

```
from numpy.linalg import inv
xval = np.dot(np.dot( inv( np.dot( A.T,A ) ), A.T), y)
print("xval = \n", xval)
print("Slope: ",  xval[0])
print("Intercept: ", xval[1])
```

#### Step 6: How does the predicted line looks like

```
pred_line = np.dot(A,xval)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.rcParams['axes.facecolor'] = 'yellow'
plt.grid(c='grey')
plt.scatter(x, y)
plt.plot(x, pred_line,'r')
plt.axis('equal')
# plt.xlabel('X')
# plt.ylabel('Y')
#plt.title('Positively Correlated')
plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)
plt.title('Linear Least Square Regression',fontsize=30)
#plt.grid()
#plt.legend(loc='upper right', fontsize=30)
plt.show()
```

## Symmetric Matrix and its Eigenvectors

### Step1: Defining Symmetric Matrix

```
import numpy as np
A = np.array(([[1,2],[3,4]]))
print("A = \n", A)
S = np.dot(A,A.T)
print("S = \n", S)
```

### Step 2: Finding Eigenvectors and Eigenvalues

```
from numpy import linalg as LA
w, v = LA.eig(S)
#w is the eigenvalue
#v is the eigenvectors corresponding to w
print("Eigenvalue = \n", w)
print("Eigenvectors = \n ", v)
# Columns of v are the eigenvectors
```

### Step 3: Orthoganality of  Eigenvectors of Symmetric Matrix

```
np.dot(v.T,v)
```

### Step 4: Observation
**For an orthonormal matrix V, its inverse is its transpose ($V^{-1} = V^{T} $). Therefore $V^{T}V = I$**

## Principal Component Analysis (PCA)

### Step 1: Creating Dataset

```
mean = (5, 4)
cov = [[2, 3], [3, 5]]

Data =  np.random.multivariate_normal(mean, cov, 5000)
print("Data shape = ", Data.shape)
```

### Step 2: Plot Original Data

```
plt.figure(figsize=(10,10))
plt.rcParams['axes.facecolor'] = 'yellow'
plt.grid(c='grey')
plt.scatter(Data[:,0], Data[:,1])

plt.axis('equal')
# plt.xlabel('X')
# plt.ylabel('Y')
#plt.title('Positively Correlated')
plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)
plt.title('Original data',fontsize=30)
#plt.grid()
#plt.legend(loc='upper right', fontsize=30)
plt.show()
```

### Step 2: Subtracting mean from data

```
mean = np.sum(Data, axis = 0, keepdims = True)/Data.shape[0]
```

```
print("mean of data", mean)
```

```
Data_mean = Data - mean
```

### Step 3: Plotting Mean subtracted data

```
plt.figure(figsize=(10,10))
plt.rcParams['axes.facecolor'] = 'yellow'
plt.grid(c='grey')
plt.scatter(Data_mean[:,0], Data_mean[:,1])

plt.axis('equal')
# plt.xlabel('X')
# plt.ylabel('Y')
#plt.title('Positively Correlated')
plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)
plt.title('Mean Subtracted Data',fontsize=30)
#plt.grid()
#plt.legend(loc='upper right', fontsize=30)
plt.show()
```

### Step 3: Covariance matrix

```
### Since we are having only two features. The covariance matrix is a 2x2 matrix

X = Data_mean.T
cov_data_mean = (1/Data_mean.shape[0]) * np.dot(X, X.T)
```

```
print("Variance - Covariance Matrix of Mean Subtracted Data = \n", cov_data_mean)
```

### Step 3: Eigenvalues and Eigenvectors of variance -covariance matrix

```
from numpy import linalg as LA
eig_value, eig_vector = LA.eig(cov_data_mean)
#w is the eigenvalue
#v is the eigenvectors corresponding to w
print("Eigenvalue = \n", eig_value)
print("Eigenvectors = \n ", eig_vector)
# Columns of v are the eigenvectors
```

### Step 4: Arranging Eigenvectors in the order of maximum eigenvalue

```
rearranged_eigenvectors = np.zeros((eig_vector.shape))
if eig_value[1] > eig_value[0]:
  rearranged_eigenvectors[:,0] = eig_vector[:,1]
  rearranged_eigenvectors[:,1] = eig_vector[:,0]
elif eig_value[1] < eig_value[0]:
  rearranged_eigenvectors[:,0] = eig_vector[:,0]
  rearranged_eigenvectors[:,1] = eig_vector[:,1]
```

```
print("Rearranged Eigenvectors based on maximum eigenvalue = \n", rearranged_eigenvectors)
```

### Step 5: Key Observations

```

print("Notice the L2 norm of eigenvectors is:", LA.norm(rearranged_eigenvectors[:,0],2))

print("Notice the L2 norm of eigenvectors is:", LA.norm(rearranged_eigenvectors[:,1],2))

print("Notice eigenvectors are orthonormal so its inverse is its transpose: \n", np.dot(rearranged_eigenvectors.T, rearranged_eigenvectors.T ))
```

### Step 6: Plotting the new basis

```

p1  = (rearranged_eigenvectors[1,0]/rearranged_eigenvectors[0,0]) *np.arange(0,10,1) # First Principal Direction
p2  = (rearranged_eigenvectors[1,1]/rearranged_eigenvectors[0,1]) * np.arange(0,10,1) # Second Principal Direction
plt.figure(figsize=(10,10))
plt.rcParams['axes.facecolor'] = 'yellow'
plt.grid(c='grey')
plt.scatter(Data_mean[:,0], Data_mean[:,1])

plt.plot(p1, 'r', label='First Principal Component')
plt.plot(p2, 'b', label='Second Principal Component')
plt.axis('equal')
# plt.xlabel('X')
# plt.ylabel('Y')
#plt.title('Positively Correlated')
plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)
plt.title('Mean Subtracted Data',fontsize=30)
#plt.grid()
plt.legend(loc='upper right', fontsize=20)
plt.show()
```

### Step 6:  Change of Basis

$Y = PData_{mean}$, $P$ is the new basis where the each row of P contains eigenvectors of covariance matrix of mean subtracted data. 

```
principal_component = rearranged_eigenvectors.T
Data_mean_transpose = Data_mean.T
### Change of Basis
Data_pca = np.dot(principal_component, Data_mean_transpose)
```

```
P[:,0]
```

### Step 7: Plotting Data after doing PCA

```


P = np.dot(principal_component, principal_component.T)


origin = [0], [0] # origin point

plt.figure(figsize=(10,10))
plt.rcParams['axes.facecolor'] = 'yellow'
plt.grid(c='grey')
plt.scatter(Data_pca[0,:], Data_pca[1,:])
plt.quiver(*origin, P[:,0], P[:,1], color=['r','b','g'], scale=5, label='First Principal Component')



plt.axis('equal')
# plt.xlabel('X')
# plt.ylabel('Y')
#plt.title('Positively Correlated')
plt.grid(True)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)
plt.title('Data after PCA',fontsize=30)
#plt.grid()
plt.legend(loc='upper right', fontsize=20)
plt.show()
```

```
import cv2
import numpy as np
from matplotlib import pyplot as plt
import scipy
import scipy.ndimage
import matplotlib.pyplot as plt 
import numpy as np 
from PIL import Image

# IMPORTING IMAGE USING SCIPY AND TAKING R,G,B COMPONENTS
a = cv2.imread('/content/drive/My Drive/PhD/Presentation/School of AI/pca/download.jpeg')
#a = scipy.ndimage.imread("/content/drive/My Drive/PhD/Presentation/School of AI/2/download.jpeg")
a_np = np.array(a)
a_r = a_np[:,:,0]
a_g = a_np[:,:,1]
a_b = a_np[:,:,2]

def comp_2d(image_2d, numpc): # FUNCTION FOR RECONSTRUCTING 2D MATRIX USING PCA
	mean_sub_im = image_2d - np.mean(image_2d , axis = None)
	eig_val, eig_vec = np.linalg.eigh(np.cov(mean_sub_im)) # USING "eigh", SO THAT PROPRTIES OF HERMITIAN MATRIX CAN BE USED
	p = np.size(eig_vec, axis =None)
	idx = np.argsort(eig_val)
	idx = idx[::-1]
	eig_vec = eig_vec[:,idx]
	eig_val = eig_val[idx]
	#numpc = 2 # THIS IS NUMBER OF PRINCIPAL COMPONENTS, YOU CAN CHANGE IT AND SEE RESULTS
	if numpc <p or numpc >0:
		eig_vec = eig_vec[:, range(numpc)]
	score = np.dot(eig_vec.T, mean_sub_im)
	recon = np.dot(eig_vec, score) + np.mean(image_2d, axis = None).T # SOME NORMALIZATION CAN BE USED TO MAKE IMAGE QUALITY BETTER
	recon_img_mat = np.uint8(np.absolute(recon)) # TO CONTROL COMPLEX EIGENVALUES
	return recon_img_mat
for numpc in np.arange(2,50,10):
  a_r_recon, a_g_recon, a_b_recon = comp_2d(a_r, numpc), comp_2d(a_g, numpc), comp_2d(a_b, numpc) # RECONSTRUCTING R,G,B COMPONENTS SEPARATELY
  recon_color_img = np.dstack((a_r_recon, a_g_recon, a_b_recon)) # COMBINING R.G,B COMPONENTS TO PRODUCE COLOR IMAGE
  recon_color_img = Image.fromarray(recon_color_img)
  #recon_color_img.show()
  plt.figure()
  plt.imshow(recon_color_img)
```
